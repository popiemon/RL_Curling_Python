```python
import pandas as pd
import pickle
import numpy as np
```

```python
with open("data/stones_map/team0_202312091806.pickle", mode="rb") as f:
    team0map = pickle.load(f)
```

```python
team0map
```

```python
MAPSIZE_X = 2.4 # (m)
MAPSIZE_Y = 43.8 # (m)
```

```python
N = 200

def matrix(x, y) -> tuple:
    numx = N*(MAPSIZE_X+x)/(2*MAPSIZE_X)
    numy = N*y/MAPSIZE_Y
    return (numx, numy)
```

```python
map_no = 11

xs = team0map[map_no]["x"].values
ys = team0map[map_no]["y"].values

team0map[map_no]
```

```python
mat = np.zeros((N, N))
for x, y in zip(xs, ys):
    num = matrix(x, y)
    if np.isnan(x):
        pass
    else:
        mat[int(num[0]), int(num[1])] = 100 # stone value
```
